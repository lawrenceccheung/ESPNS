# Test of Laguerre Polynomial Multiplication
$$
L_r(x) L_s(x) = \sum_t A_{rst} L_t(x)
$$
where
$$
A_{rst} = \sum_{p=0}^r \sum_{q=0}^s (-1)^{p+q+t} { r \choose p }
{ s \choose q }  { {p+q}\choose{p} }  { {p+q}\choose t }
$$

# New Section

```python
from math import *
from pylab import *

# Define NchooseK
def nchoosek(n,k):
    if (n-k>=0):
        return factorial(n)/factorial(k)/factorial(n-k)
    else:
        return 0
    
def Arst(r,s,t):
    sum=0
    for p in range(r+1):
        for q in range(s+1):
            coeff=(-1)**(p+q+t)
            sum = sum + coeff*nchoosek(r,p)*nchoosek(s,q)*nchoosek(p+q,p)*nchoosek(p+q,t)
    return sum
```
